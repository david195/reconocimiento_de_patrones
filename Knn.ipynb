### Knn aplicado a el conjunto Iris

Se aplicara el algortimo de clasificaci√≥n knn al conjunto de datos
iris del repositorio de la UCI.

```python
#Carga de bibliotecas
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
from statistics import mode
data = load_iris().data
target = load_iris().target
```

```python
ntest = 0.2
nclass = 3
ex_cl = 50
```

```python
index = np.concatenate([np.arange(0, int((1-ntest)*ex_cl)), np.arange(ex_cl, ex_cl+int((1-ntest)*ex_cl)), np.arange(2*ex_cl, 2*ex_cl+int((1-ntest)*ex_cl))])
xtrain = np.array(data[index])
ytrain = target[index]
print("Conjunto de entrenamiento:\n {}\n".format(xtrain))

index = np.concatenate([np.arange(int((1-ntest)*ex_cl), ex_cl), np.arange(ex_cl+int((1-ntest)*ex_cl) , 2*ex_cl), np.arange(2*ex_cl+int((1-ntest)*ex_cl),3*ex_cl)])
xtest = np.array(data[index])
ytest = target[index]
print("Conjunto de prueba:\n {}\n".format(xtest))
```

```python
print("Etiquetas del conjunto de entrenamiento: {}\n".format(ytrain))

print("Etiquetas del conjunto de prueba: {}\n".format(ytest))
```

```python
def knn(x, train, target, k):
    d = pow((train - x),2) 
    d = np.sqrt(d.sum(axis=1))
    print(d)
    dt = np.array([[di, yi] for di, yi in zip(d, target)])
    return(np.sort(dt, axis=0)[0:k,:])
```

```python
S = []
k = 11;
for i in range(0,xtest.shape[0]):
    d = knn(xtest[i], xtrain, ytrain, k)
    S = np.append(S,mode(d[:,1]))
msal = (S==ytest)*1
accuracy = msal.sum()/(nclass*ntest*ex_cl)
print("accuracy: {}".format(accuracy))
```
